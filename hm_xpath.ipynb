```python
from pprint import pprint
from lxml import html
import requests
from datetime import datetime

header = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36'}
main_link = 'https://news.mail.ru/'
response = requests.get(main_link, headers=header)
dom = html.fromstring(response.text)
```

```python
mail_news_block = dom.xpath('//table[contains(@class,"daynews__inner")]//a')

news1 = []

for block in mail_news_block:
    
    news_upd = {}
    name = block.xpath('.//span[contains(@class,"photo__title_new_hidden")]/text()')[0].replace('\xa0',' ')
    link = block.xpath('./@href')[0]
    date = dom.xpath('//body/@data-now')[0]
    
    news_upd['source'] = main_link
    news_upd['name'] = name
    news_upd['link'] = main_link + link
    news_upd['upd_at'] = datetime.utcfromtimestamp(int(date)).strftime('%Y-%m-%d %H:%M:%S')
    
    news1.append(news_upd)
    
pprint(news1)
```

```python
header = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36'}
main_link = 'https://lenta.ru/'
response = requests.get(main_link, headers=header)
dom = html.fromstring(response.text)
```

```python
lenta_news_block = dom.xpath('//div[contains(@class,"b-yellow-box__wrap")]//a')

news2 = []

for block in lenta_news_block:
    
    news_upd = {}
    name = block.xpath('./text()')[0].replace('\xa0',' ')
    link = block.xpath('./@href')[0]
    date = dom.xpath('//time[@class]/text()')[0]
    
    news_upd['source'] = main_link
    news_upd['name'] = name
    news_upd['link'] = main_link + link
    news_upd['upd_at'] = datetime.strftime(datetime.now(), "%Y.%m.%d") + " " + date
    
    news2.append(news_upd)
    
pprint(news2)
```

```python
header = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36'}
main_link = 'https://yandex.ru'
folder = '/news'
response = requests.get(main_link + folder, headers=header)
dom = html.fromstring(response.text)
```

```python
yandex_news_block = dom.xpath('//div[contains(@aria-label,"Главное")]//h2')

news3 = []

for block in yandex_news_block:
    
    news_upd = {}
    name = block.xpath('.//a/text()')[0]
    link = block.xpath('.//a/@href')[0]
    date = dom.xpath('//div[contains(@class,"story__date")]/text()')[0].split()[1]
    
    news_upd['source'] = main_link
    news_upd['name'] = name
    news_upd['link'] = main_link + link
    news_upd['upd_at'] = datetime.strftime(datetime.now(), "%Y.%m.%d") + " " + date
    
    news3.append(news_upd)
    
pprint(news3)
```

```python
result = news1 + news2 + news3
pprint(result)
```
